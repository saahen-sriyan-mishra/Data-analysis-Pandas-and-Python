#### Year should be converted into int instead of characters
To remove any cell that is not int, and replace it with zero.

```python
# Replace non-numeric values with an empty string
df['year'] = df['year'].replace(r'\D', '', regex=True)

# Convert year col. to numeric type, errors for non-convertible values to NaN
df['year'] = pd.to_numeric(df['year'], downcast='integer', errors='coerce')

# Fill NaN values with 0
df['year'] = df['year'].fillna(0)

# Convert 'year' column to integer type
df['year'] = df['year'].astype(int)
```

#### Drop unwanted column

```python
new_df = df.drop(['movie_id', 'director_id', 'star_id', 'description'], axis=1)
```

#### No need for extra char. in star anD director col.

```python
new_df['director'] = new_df['director'].str.replace('\n', '')
new_df['star'] = new_df['star'].str.replace('\n', '')
```

#### Drop row if runtime = 0

```python
new_df = new_df.dropna(subset=['runtime'])
```

```python
new_df
```

#### Converting runtime to date time format

```python
#Change string type in runtime to be int
new_df.loc[:, 'runtime'] = new_df['runtime'].astype(str)
new_df.loc[:, 'runtime'] = new_df['runtime'].str.replace('min', '')
new_df.loc[:, 'runtime'] = new_df['runtime'].str.replace(',', '').astype(int)
# using .loc with : since pandas want to save a copy warning be avoided 
```

```python
# Make it into datetime format
new_df.loc[:, 'runtime'] = pd.to_timedelta(new_df['runtime'], unit='m')
```

**Get movie with the highest rating for each year.**

```python
# Create an empty DataFrame to store highest-rated movies
highest_rated_movies_df = pd.DataFrame(columns=new_df.columns)

# Iterate through each unique year
for year in new_df['year'].unique():
    # Filter by year
    df_year = new_df[new_df['year'] == year]
    
    if not df_year.empty:  # Check if the resulting DataFrame is not empty
        # Sort by ratings in descending order
        df_year_sorted = df_year.sort_values(by='rating', ascending=False)
        # Get the highest rating
        highest_rating = df_year_sorted.iloc[0]['rating']
        # Filter movies with the highest rating for the year
        highest_rated_movies_year = df_year_sorted[df_year_sorted['rating'] == highest_rating]
        # Append the filtered DataFrame to the main DataFrame
        highest_rated_movies_df = pd.concat([highest_rated_movies_df, highest_rated_movies_year])

# Reset index and sort by year
highest_rated_movies_df = highest_rated_movies_df.reset_index(drop=True).sort_values(by='year', ascending=False)

# Display the highest-rated movies
highest_rated_movies_df[['movie_name','year','rating','genre']].reset_index(drop=True)
```

**Get average movie duration for each year.**

```python
avg_duration_movies = []

# Iterate each unique year 
for year in new_df['year'].unique():
    # Filter by year
    if year < 2024: 
        df_year = new_df[new_df['year'] == year]
    
        if not df_year.empty:  # Check empty
            df_year_sorted = df_year.sort_values(by='runtime', ascending=False)
            
            total_runtime_seconds = 0  # Initialize total runtime for movies
            
            for i in range(len(df_year_sorted)):
                # Access the 'runtime' value of each row and convert to seconds
                runtime_timedelta = df_year_sorted.iloc[i]['runtime']
                total_runtime_seconds += runtime_timedelta.total_seconds()
            
            # Calculate the average runtime for movies in this year
            avg_runtime_seconds = total_runtime_seconds / len(df_year_sorted)
            
            # Convert average runtime from seconds to Timedelta
            avg_runtime_timedelta = pd.Timedelta(seconds=avg_runtime_seconds)
            
            # Append the average runtime to the list
            avg_duration_movies.append({'Year': year, 'Average Runtime': avg_runtime_timedelta})

# Create a DataFrame from the list of dictionaries
avg_duration_movies_df = pd.DataFrame(avg_duration_movies)

# Sort the DataFrame by 'year' in descending order
x5 = avg_duration_movies_df.sort_values(by='Year', ascending=False).reset_index(drop=True)

x5.reset_index(drop =True)
```

**Relationship between years and gross revenue (boxplot).**

```python
# Year ranges
year_ranges = [(None, 1950), (1951, 1960), (1961, 1970), (1971, 1980),
               (1981, 1990), (1991, 2000), (2001, 2010), (2011, 2020)]

# Set subplots
fig, axs = plt.subplots(len(year_ranges), figsize=(5, 3 * len(year_ranges)))

for i, (start_year, end_year) in enumerate(year_ranges):
    # Filter data for the current year_range
    if start_year is None:
        filtered_data = df[df['year'] <= end_year]['gross(in $)']
    else:
        filtered_data = df[(df['year'] >= start_year) & (df['year'] <= end_year)]['gross(in $)']
    
    # Drop NaN, Inf
    cleaned_data = filtered_data.dropna().replace([np.inf, -np.inf], np.nan).dropna()
    
    # Plot boxplot
    axs[i].boxplot(cleaned_data, showfliers=False)
    axs[i].set_title(f'Boxplot for {start_year if start_year is not None else "Prior to"} - {end_year} Gross')
    
plt.tight_layout()
plt.show()
```

**Relationship between runtime and average revenue (line graph).**

```python
# Runtime ranges
runtime_ranges = [(0, 20), (20, 40), (40, 60), (60, 80), (80, 100),
                  (100, 120), (120, 150), (150, 180), (180, 240), (240, 999999)]

# Convert timedelta to minutes
df['runtime_minutes'] = df['runtime'] / pd.Timedelta(minutes=1)

# Filter 
filtered_df = df[df['runtime_minutes'].notna()]  # Drop rows with NaN 
for i in range(len(runtime_ranges)):
    filtered_df.loc[(filtered_df['runtime_minutes'] >= runtime_ranges[i][0]) & 
                    (filtered_df['runtime_minutes'] < runtime_ranges[i][1]), 
                    'runtime_range'] = f"{runtime_ranges[i][0]} to {runtime_ranges[i][1]}"
    
# Get unique ranges
unique_ranges = filtered_df['runtime_range'].unique()

# Split runtime ranges into start time and finish time
start_times = [int(range_.split(' ')[0]) for range_ in unique_ranges]
finish_times = [int(range_.split(' ')[2]) if range_.split(' ')[2] != '999999' else 999999 for range_ in unique_ranges]

sorted_indices = sorted(range(len(unique_ranges)), key=lambda x: (start_times[x], finish_times[x]))

sorted_ranges = [unique_ranges[i] for i in sorted_indices]

# Set up subplots
num_cols = 3  # Number of columns for subplots
num_rows = (len(sorted_ranges) // num_cols) + (len(sorted_ranges) % num_cols)  # Calculate number of rows

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Flatten axes if necessary
if num_rows == 1:
    axes = [axes]

# Plot each runtime range on a separate subplot
for i, runtime_range in enumerate(sorted_ranges):
    start_year, finish_year = runtime_range.split(' ')[0], runtime_range.split(' ')[2]
    if finish_year == '999999':
        range_data = filtered_df[filtered_df['runtime_minutes'] >= runtime_ranges[i][0]]
        title = f'Average Rating Over the runtime of {start_year} and beyond'
    else:
        range_data = filtered_df[filtered_df['runtime_range'] == runtime_range]
        title = f'Average Rating Over the runtime of {start_year} to {finish_year}'
    ax = axes[i // num_cols][i % num_cols]
    sns.lineplot(data=range_data, x='year', y='rating', ax=ax)
    ax.set_xlim(1900, None)  # Set x-axis 
    ax.set_title(title)
    ax.set_xlabel('Year')
    ax.set_ylabel('Average Rating')

# Hide unused subplot
for i in range(len(sorted_ranges), num_rows * num_cols):
    axes[i // num_cols][i % num_cols].axis('off')
#show
plt.tight_layout()
plt.show()
```

**Realtion between Runtime to Average Gross (in $). (Bar Graph)**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Runtime ranges
runtime_ranges = [(0, 20), (20, 40), (40, 60), (60, 80), (80, 100), 
                  (100, 120), (120, 150), (150, 180), (180, 240), (240, float('inf'))]

# Convert timedelta to minutes
df['runtime_minutes'] = df['runtime'] / pd.Timedelta(minutes=1)

# Filter 
filtered_df = df[df['runtime_minutes'].notna()]  # Drop rows with NaN 
for i in range(len(runtime_ranges)):
    filtered_df.loc[(filtered_df['runtime_minutes'] >= runtime_ranges[i][0]) & 
                    (filtered_df['runtime_minutes'] < runtime_ranges[i][1]), 
                    'runtime_range'] = f"{runtime_ranges[i][0]} to {runtime_ranges[i][1]}"

# Calc. the average gross rev.
average_gross = []
for runtime_range in runtime_ranges:
    start, end = runtime_range
    if end == float('inf'):
        range_data = filtered_df[(filtered_df['runtime_minutes'] >= start)]
    else:
        range_data = filtered_df[(filtered_df['runtime_minutes'] >= start) & 
                                 (filtered_df['runtime_minutes'] < end)]
        # Convert to millions
    average_gross.append(range_data['gross(in $)'].mean() / 1e6)  

# Plot the bar graph
plt.figure(figsize=(10, 6))
bars = plt.bar(range(len(runtime_ranges)), average_gross, color='skyblue')

# Add data value 
for bar in bars:
    yval = bar.get_height()
    if pd.notna(yval) and pd.notnull(yval):  # Check if y-value is not NaN or None
        plt.text(bar.get_x() + bar.get_width()/2, yval, f"{round(yval, 2)}M", va='bottom')  # Display in millions

# x-axis labels to handle "240 and beyond" case
x_labels = [f"{start}-{end}" if end != float('inf') else f"{start} and beyond" for start, end in runtime_ranges]
plt.xticks(range(len(runtime_ranges)), x_labels, rotation=45)

plt.xlabel('Runtime Range')
plt.ylabel('Average Gross (in $) (Millions)')
plt.title('Average Gross Revenue by Movie Runtime Range')
plt.tight_layout()
plt.show()
```
